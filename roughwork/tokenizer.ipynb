```python
def flatten(nested_list):
    flattened = []
    for item in nested_list:
        if isinstance(item, (tuple, list)):
            flattened.extend(flatten(item))
        else:
            flattened.append(item)
    return flattened

flatten(['a', (1,2), [1, 2, 3]])
```

```python
def top_pair(tokens):
    counts = {}
    for pair in zip(tokens, tokens[1:]):
        counts[pair] = counts.get(pair, 0)+1

    return counts
```
